```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8')
```

```python
np.random.seed(0)
```

```python
def st_petersburg_simulation(num_games):
    total_payout = []
    for _ in range(num_games):
        num_flips = 1
        while np.random.rand() > 0.5:
            num_flips += 1
        payout = 2 ** num_flips
        total_payout.append(payout)
    return np.array(total_payout)
```

```python
# Number of games to simulate
num_games = 100000
range_number = np.arange(1, num_games+1)
# Run the simulation
game_payoff = st_petersburg_simulation(num_games)
```

```python
mean_gamge = np.cumsum(game_payoff) / range_number
```

```python
plt.figure(figsize=(12, 5))
plt.plot(range_number, mean_gamge)
```
