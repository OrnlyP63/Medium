## Discrete Urysohn operators

$$F(x_1, x_2, \ldots, x_n) = \sum_{q=1}^{2n+1}\Phi_q\left(\sum_{p=1}^n\phi_{q, p}(x_p)\right)$$

Consider:

function $f = x_1f_1 + x_2f_2$; $x_1 + x_2 = 1$

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
f_i = np.array([3, 10])
d_i = np.array([0.1, 0.9])
x_i = np.array([0, 1])
```

```python
f = f_i @ d_i
x = x_i @ d_i
```

```python
plt.plot(x_i, f_i, '-o')
plt.plot([x, x], [f, 0], '--o')
plt.ylim(0, 15)
```

Regression model: $$y = \sum_{j=1}^nf_i(x_j)$$

converts components $x_1, x_2, \ldots$ of vector $X$ into scalar $y$ by function $f_1, f_2, \ldots$


Two piecewise linear functions $f_1$ and $f_2$ with ordinates $f_{1, 1}, f_{1, 2}, \ldots, f_{2, 1}, f_{2, 2}, \ldots$

We can make a block vector with unknown ordinates: 
$$F^\top = 
\begin{bmatrix}
f_{1, 1} & f_{1, 2} & f_{1, 3} & f_{1, 4} & f_{1, 5} \\
f_{2, 1} & f_{2, 2} & f_{2, 3} & f_{2, 4} & f_{2, 5} \\
\end{bmatrix}$$

and a block vector with known values: 

$$X^\top = 
\begin{bmatrix}
0 & 0.1 & 0.9 & 0 & 0 \\
0 & 0.9 & 0.1 & 0 & 0 \\
\end{bmatrix}$$



```python
f_i = np.array([3, 6, 7, 9, 8])
d_i = np.array([0.0, 0.1, 0.9, 0.0, 0.0])
x_i = np.linspace(0, 1, 5)
```

```python
f = f_i @ d_i
x = x_i @ d_i
```

```python
plt.plot(x_i, f_i, '-o')
plt.plot([x, x], [f, 0], '--o')
plt.ylim(0, 15)
plt.ylabel('$f$');
```
