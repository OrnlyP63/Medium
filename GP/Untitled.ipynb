# Stochastic Process

To simulate Brownian motion (Wiener process) over time $t$ in 1 dimension $d$ and moving the particle over a certain amount of time $\Delta t$ with a random distance $\Delta d$ from previous position.

```python
#1d simulation Brawnian motion
time = 1
nb_step = 100
delta_time = time / nb_step

nb_motion = 10
mean = 0.
std = np.sqrt(delta_time)

random_data = np.random.normal(mean, std, (nb_motion, nb_step))
distance = np.cumsum(random_data, axis=1)
```

```python
plt.figure(figsize=(6,4))
t = np.arange(0, time, delta_time)
for i in range(nb_motion):
    plt.plot(t, distance[i])
plt.title('Bronian motion')
plt.show()
```

# Kernel function
- Covariance function as prior

$$k(x_i, x_j) = exp(-\frac{1}{2\sigma^2}||x_i-x_j||^2)$$

```python
def kernel(xi, xj):
    sq_norm = -0.5 * sp.spatial.distance.cdist(xi, xj, 'sqeuclidean')
    return np.exp(sq_norm)
```

```python
nb_samples = 50
nb_func = 5

X = np.expand_dims(np.linspace(-4, 4, nb_samples), 1)
sigma = kernel(X, X)
```

```python
ys = np.random.multivariate_normal(
    mean=np.zeros(nb_samples), 
    cov=sigma, 
    size=nb_func)
```

```python
plt.figure(figsize=(6, 4))
for i in range(nb_func):
    plt.plot(X, ys[i], marker='o', markersize=3)
plt.title('5 random function')
```

- Predictions from posterior

$$\mu_{2|1} = \Sigma_{21}\Sigma_{11}^{-1}y_1 = (\Sigma_{11}\Sigma_{21}^{-1})^Ty_1$$

$$\Sigma_{2|1} = \Sigma_{22} - \Sigma_{21}\Sigma_{11}^{-1}\Sigma_{12} = \Sigma_{22} - (\Sigma_{11}\Sigma_{21}^{-1})^T\Sigma_{12}$$

```python
def GP(X1, y1, X2):
    sigma11 = kernel(X1, X1)
    sigma12 = kernel(X1, X2)
    sigma22 = kernel(X2, X2)
    s = sp.linalg.solve(sigma11, sigma12, assume_a='pos').T
    
    mu2 = s @ y1
    sigma2 = sigma22 - (s @ sigma12)
    
    return mu2, sigma2
```

```python
sin_function = lambda x: (np.sin(x)).flatten()

n_train = 8
n_test = 75
n_func = 5
domain = (-6, 6)

X1 = np.random.uniform(domain[0]+2, domain[1]-2, size=(n_train, 1))
y1 = sin_function(X1)

X2 = np.linspace(domain[0], domain[1], n_test).reshape(-1, 1)
```

```python
mu2, sigma2 = GP(X1, y1, X2)
c2 = np.sqrt(np.diag(sigma2))
```

```python
y2 = np.random.multivariate_normal(mean=mu2, cov=sigma2, size=n_func)
```

```python
fig, (ax1, ax2) = plt.subplots(
    nrows=2, ncols=1, figsize=(8, 6))

# Plot the distribution of the function (mean, covariance)
ax1.plot(X2, sin_function(X2), 'b--', label='$sin(x)$')
ax1.fill_between(X2.flat, mu2-2*c2, mu2+2*c2, color='red', 
                 alpha=0.15, label='$2 \sigma_{2|1}$')

ax1.plot(X2, mu2, 'r-', lw=2, label='$\mu_{2|1}$')
ax1.plot(X1, y1, 'ko', linewidth=2, label='$(x_1, y_1)$')
ax1.set_xlabel('$x$', fontsize=13)
ax1.set_ylabel('$y$', fontsize=13)
ax1.set_title('Distribution of posterior and prior data.')
ax1.axis([domain[0], domain[1], -3, 3])
ax1.legend()

ax2.plot(X2, y2.T, '-')
ax2.set_xlabel('$x$', fontsize=13)
ax2.set_ylabel('$y$', fontsize=13)
ax2.set_title('5 different function realizations from posterior')
ax1.axis([domain[0], domain[1], -3, 3])
ax2.set_xlim([-6, 6])
plt.tight_layout()
plt.show()
```
