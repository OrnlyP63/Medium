```python
import matplotlib.pyplot as plt
import yfinance as yf
import numpy as np
```

```python
ticker = "^GSPC"

# Download the data
spx_data = yf.download(ticker, interval="1d")
```

```python
spx_data = spx_data[spx_data.index.year > 2018]
```

```python
ret = np.log(spx_data['Close']).diff()
```

```python
ret.plot()
```
