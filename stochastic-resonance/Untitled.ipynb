## Resting State

$$V(x) = \frac{b}{4}x^4 - \frac{a}{2}x^2$$

## Activated state

$$V(x, t) = V(x) - Ax\sin(\omega t)$$

```python
activate_state = lambda x,a,b,t,A : (b/4)*x**4 - (a/2)*x**2 + A*x*np.sin(0.001*t)
```

```python
a, b = 1, 1
A = 0.35
T = 10
x = np.arange(-4, 4.1, 0.1)
x0 = -2.0

dt = 0.01
sigma = 0.2

number_of_samples = int(T / dt)

results = []

for t in range((number_of_samples)):
    y = activate_state(x0, a, b, t, A) 
    
    results.append((x0, y))
    
    dx = -diff_activate_state(x0, a, b, t, A) * dt + sigma * np.random.normal()
    x0 = x0 + dx
```

```python
results = np.array(results)
```

```python
plt.plot(range(number_of_samples), results[:, 0]);
```
