## Resting State

$$V(x) = \frac{b}{4}x^4 - \frac{a}{2}x^2$$

## Activated state

$$V(x, t) = V(x) - Ax\cos(\omega t)$$

```python
activate_state = lambda x,a,b,t,A : (b/4) * x ** 4 - (a/2) * x ** 2 - A * x * np.cos(0.001*t)
neg_diff_activate_state = lambda x,a,b,t,A : - b * x ** 3 + a * x + A * np.cos(0.001*t)
```

```python
a, b = 1, 1
A = 0.35
T = 100
x = 0.0
y = activate_state(x, a, b, 0, A) 

dt = 0.1
sigma = 0.27

number_of_samples = int(T / dt)

results = []
# x = np.zeros(num_steps)
# v = np.zeros(num_steps)
```

```python
for t in range(number_of_samples):
    results.append((x, y))
    dx = neg_diff_activate_state(x, a, b, t, A) * dt + sigma * np.random.normal()
    
    x = x + dx
    y = activate_state(x, a, b, 0, A) 
```

```python
results = np.array(results)
```
