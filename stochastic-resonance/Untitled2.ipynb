```python
import numpy as np
import matplotlib.pyplot as plt
# import sympy
# plt.style.use('seaborn-pastel')
```

## Resting State

$$V(x) = \frac{b}{4}x^4 - \frac{a}{2}x^2$$

## Activated state

$$V(x, t) = V(x) - Ax\cos(\omega t)$$

```python
activate_state = lambda x,a,b,t,A : (b/4) * x ** 4 - (a/2) * x ** 2 - A * x * np.cos(2*np.pi*0.001*t)
diff_activate_state = lambda x,a,b,t,A : b * x ** 3 - a * x - A * np.cos(2*np.pi*0.001*t)
```

```python
for sigma in  np.linspace(0.01, 1, 100):
    a, b = 1, 1
    A = 0.1
    T = 100
    x = np.arange(-4, 4.1, 0.1)
    x = -2.0
    y = activate_state(x, a, b, 0, A) 

    dt = 0.001
    # sigma = 0.03

    number_of_samples = int(T / dt)

    results = []

    for t in range(number_of_samples):
        
        dx = -diff_activate_state(x, a, b, t, A) * dt + sigma * np.random.normal()

        x = x + dx
        y = activate_state(x, a, b, 0, A) 
        results.append((x, y))
        
    results = np.array(results)
    plt.figure(figsize=(10, 4))
    plt.plot(range(number_of_samples), results[:, 0])
    plt.show()
```
