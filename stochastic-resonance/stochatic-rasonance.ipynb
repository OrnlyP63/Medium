```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-pastel')
```

## Resting State

$$V(x) = \frac{b}{4}x^4 - \frac{a}{2}x^2$$

```python
x = np.arange(-3, 3.1, 0.1)

# resting state
a, b = 2, 0.5

rest_state = lambda x,a,b : (b/4)*x**4 - (a/2)*x**2
plt.plot(x, rest_state(x, a, b));
```

## Activated state

$$V(x, t) = V(x) - Ax\sin(\omega t)$$

```python
from matplotlib import animation
from matplotlib import rc
rc('animation', html='html5')
```

```python
fig = plt.figure()
ax = plt.axes(xlim=(-4, 4), ylim=(-5, 2))
line, = ax.plot([], [], lw=3)

def init():
    line.set_data([], [])
    return line,

# activated state
activate_state = lambda x,a,b,t : (b/4)*x**4 - (a/2)*x**2 + x*np.sin(0.03*t)
a, b = 2, 0.5

def animate(t):
    x = np.arange(-4, 4.1, 0.1)
    y = activate_state(x, a, b, t)
    line.set_data(x, y)
    return line,
```

```python
anim = animation.FuncAnimation(fig, animate, frames=1000, interval=10, blit=True)
```
