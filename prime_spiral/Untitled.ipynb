```python
import math
import sympy
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
plt.style.use('dark_background')
```

```python
def get_coordinate(num):
    return num * np.cos(num), num * np.sin(num)
```

```python
x = np.arange(1, 10)
get_coordinate(x)
```

```python
def create_plot(nums, figsize=8, s=None, show_annot=False):
    nums = np.array(list(nums))
    x, y = get_coordinate(nums)
    plt.figure(figsize=(figsize, figsize))
    plt.axis("off")
    plt.scatter(x, y, s=s)
    # plt.show()
```

```python
primes = sympy.primerange(0, 20000)
create_plot(primes, s=1)
```

```python
primes = sympy.primerange(0, 400000)
create_plot(primes, s=0.15)
```

```python
print(list(sympy.primerange(0, 100)))
```

```python
def twin_primes(n):
    primes = np.array(list(sympy.primerange(0, n)))

    mask = (primes[1:] - primes[:-1]) == 2

    new_mask = np.append(mask, [False], axis=0) | np.append([False], mask, axis=0)

    return primes[new_mask]
```

```python
primes = twin_primes(400000)
create_plot(primes, s=1)
```

```python
primes = twin_primes(2**15)
create_plot(primes, s=0.15)
plt.show()
```

```python
figsize = 25 
s = 0.2

plt.figure(figsize=(figsize, figsize))
plt.suptitle(r'$\sigma_2(n) = 5 + 2n + n^2$', fontsize=32, y=0.95)
# plt.subplots_adjust(hspace=0.2)
for i, p in enumerate(np.arange(20, 29)):
    
    primes = twin_primes(int(1.618 ** p ))
    x, y = get_coordinate(primes)
    ax = plt.subplot(3, 3, i+1)
    ax.scatter(x, y, s=s)
    ax.axis("off")
plt.savefig('twin_primes3.png', dpi=300)
```
