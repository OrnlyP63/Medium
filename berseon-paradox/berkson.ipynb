## บทพิสูจน์ทางคณิตศาสตร์ของ Berkson's Paradox (Collider Bias)

ปรากฏการณ์นี้สามารถพิสูจน์ได้ 2 รูปแบบหลัก คือแบบตรรกะบูลีน (ตามเปเปอร์ต้นฉบับ) และแบบความแปรปรวนร่วม (สำหรับใช้อธิบายระบบคัดกรองแบบให้คะแนนรวม)

### 1. บทพิสูจน์แบบตรรกะบูลีน (Discrete Proof)
เป็นการพิสูจน์ว่าทำไมตัวแปรที่เป็นอิสระต่อกัน ถึงกลายมามีความสัมพันธ์เชิงลบเมื่อถูกคัดกรองผ่านเงื่อนไข "อย่างใดอย่างหนึ่ง"

**กำหนดให้:**
* $X$ และ $Y$ เป็นเหตุการณ์แบบสุ่มที่เป็นอิสระต่อกัน (Independent) โดยสมบูรณ์
* $P(X=1 \mid Y=1) = P(X=1)$
* ให้ความน่าจะเป็นของแต่ละเหตุการณ์คือ $P(X=1) = p$ และ $P(Y=1) = q$ โดยที่ $0 < p, q < 1$

**เงื่อนไขการคัดกรอง (The Collider):**
สมมติให้ระบบ HR จะคัดเลือกคน ($Z=1$) ก็ต่อเมื่อเขามีทักษะ $X$ หรือ $Y$ อย่างน้อยหนึ่งอย่าง
$$Z = X \lor Y$$

**การพิสูจน์:**
เราต้องการหาความน่าจะเป็นที่ผู้สมัครจะมีทักษะ $X$ ภายใต้เงื่อนไขว่าเขาถูกคัดเลือกเข้ามาแล้ว ($Z=1$) โดยแบ่งเป็น 2 กรณี:

**กรณีที่ 1: ผู้สมัครมีทักษะ $Y$ ($Y=1$)**
เมื่อ $Y=1$ เงื่อนไข $Z=1$ จะเป็นจริงโดยอัตโนมัติ ดังนั้นเงื่อนไข $Z$ จึงไม่ได้เพิ่มข้อมูลใหม่:
$$P(X=1 \mid Y=1, Z=1) = P(X=1 \mid Y=1) = P(X=1) = p$$

**กรณีที่ 2: ผู้สมัครไม่มีทักษะ $Y$ ($Y=0$)**
เพื่อให้เขาผ่านการคัดเลือก ($Z=1$) ได้ในขณะที่ไม่มีทักษะ $Y$ วิธีเดียวคือเขาบังคับจะต้องมีทักษะ $X$:
$$P(X=1 \mid Y=0, Z=1) = 1$$

**บทสรุปทางคณิตศาสตร์:**
เนื่องจากกำหนดไว้ตั้งแต่ต้นว่าประชากรทั่วไปไม่ได้เก่งสมบูรณ์แบบ ($p < 1$) เราจึงได้อสมการ:
$$P(X=1 \mid Y=1, Z=1) < P(X=1 \mid Y=0, Z=1)$$
**ความหมาย:** ในกลุ่มคนที่ผ่านการคัดกรอง การทราบว่าเขามีทักษะด้านหนึ่ง ($Y$) จะทำให้โอกาสที่เขาจะมีอีกทักษะหนึ่ง ($X$) **ลดลง** ทันที

---

### 2. บทพิสูจน์แบบความแปรปรวนร่วม (Continuous / Covariance Proof)
เป็นการพิสูจน์ที่สะท้อนการทำงานของอัลกอริทึม ATS (ระบบคัดกรองเรซูเม่) ที่ให้คะแนนผู้สมัครแบบบวกทบกัน (Additive Scoring)

**กำหนดให้:**
* $X$ (ทักษะทางวิศวกรรม) และ $Y$ (ทักษะการทำเรซูเม่) เป็นตัวแปรต่อเนื่องที่เป็นอิสระต่อกัน
* ดังนั้น ความแปรปรวนร่วมจะเท่ากับศูนย์: $Cov(X, Y) = 0$
* เกณฑ์ของ HR คือผลรวมคะแนน: $S = X + Y$

**การพิสูจน์:**
สมมติให้ HR รับเฉพาะคนที่ได้คะแนนรวม $s$ พอดี ($X + Y = s$) เราต้องการหาความสัมพันธ์ระหว่าง $X$ และ $Y$ ภายใต้เงื่อนไขนี้ นั่นคือหาค่า $Cov(X, Y \mid X+Y=s)$

จากเงื่อนไข $X + Y = s$ เราจะได้ว่า $Y = s - X$
นำกลับไปแทนค่าในสมการ Covariance:
$$Cov(X, Y \mid X+Y=s) = Cov(X, s - X)$$

กระจายพจน์ตามคุณสมบัติพื้นฐานของ Covariance ($Cov(A, B+C) = Cov(A, B) + Cov(A, C)$):
$$Cov(X, s - X) = Cov(X, s) - Cov(X, X)$$

เนื่องจาก $s$ เป็นค่าคงที่ ความแปรปรวนร่วมกับค่าคงที่จึงมีค่าเป็นศูนย์ ($Cov(X, s) = 0$) และ $Cov(X, X)$ ก็คือความแปรปรวน ($Var(X)$):
$$Cov(X, Y \mid X+Y=s) = -Var(X)$$

**บทสรุปทางคณิตศาสตร์:**
เนื่องจากความแปรปรวน $Var(X)$ ต้องมีค่าเป็นบวก (Positive) เสมอ ผลลัพธ์ที่ได้จึง **ติดลบเสมอ** นี่คือการพิสูจน์ว่า การนำตัวแปรสองตัวมารวมกันเพื่อสร้างเกณฑ์คัดกรอง จะเป็นการบีบให้ตัวแปรทั้งสองเกิดความสัมพันธ์เชิงลบ (Negative Correlation) เสมอ ทำให้องค์กรได้พนักงานที่ "เก่งอย่างใดอย่างหนึ่ง แต่ไม่เก่งอีกอย่าง" เข้ามาสะสมตัวจนเกิดความเสี่ยง

---
**อ้างอิงเปเปอร์ทฤษฎีรากฐาน (Berkson's Paradox):**
* Berkson, J. (1946). *Limitations of the Application of Fourfold Table Analysis to Hospital Data*. Biometrics Bulletin, 2(3), 47-53. 
* **ลิงก์:** [https://doi.org/10.2307/3002000](https://doi.org/10.2307/3002000)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from scipy import stats
```

```python
# 1. ตั้งค่าเริ่มต้นประชากรผู้สมัครงาน
np.random.seed(42)
n_candidates = 100000

# สร้างตัวแปร X (True Skill) และ Y (Resume Hacking Skill) 
# ให้เป็นอิสระต่อกันโดยสมบูรณ์ในธรรมชาติ (Correlation เข้าใกล้ 0)
# สมมติให้คะแนนเต็ม 100, ค่าเฉลี่ย 50, ส่วนเบี่ยงเบนมาตรฐาน 15
X = np.random.normal(50, 15, n_candidates)
Y = np.random.normal(50, 15, n_candidates)

# ตัวชี้วัดรวมที่ HR ใช้ประเมิน (Additive Score: S = X + Y)
S = X + Y

# 2. กำหนดระดับความเข้มงวดของ HR (รับคนที่ได้คะแนนรวม Top % เท่าไหร่)
# 100 = รับทุกคน (ไม่มีการคัดกรอง), 1 = รับแค่ระดับหัวกะทิ Top 1% ของเรซูเม่
selection_percentiles = [100, 50, 20, 10, 5, 1]

results = []

print(f"{'ความเข้มงวด HR':<20} | {'จำนวนคนที่รับ':<15} | {'Correlation (X, Y)':<20}")
print("-" * 60)

for p in selection_percentiles:
    # หาคะแนนจุดตัด (Threshold) ตาม Percentile ที่ตั้งไว้
    threshold = np.percentile(S, 100 - p)
    
    # คัดเฉพาะคนที่คะแนนรวมผ่านเกณฑ์ (The Collider Box)
    selected_mask = S >= threshold
    X_selected = X[selected_mask]
    Y_selected = Y[selected_mask]
    
    # คำนวณความสัมพันธ์ (Correlation) ในกลุ่มคนที่ผ่านเข้ารอบ
    corr, _ = pearsonr(X_selected, Y_selected)
    results.append((p, corr))
    
    label = f"รับเฉพาะ Top {p}%" if p < 100 else "รับทุกคน (ไม่มีเกณฑ์)"
    print(f"{label:<20} | {len(X_selected):<15} | {corr:>.4f}")

# 3. สร้างกราฟแสดงความพังพินาศของระบบ (The Breakdown)
plt.figure(figsize=(10, 6))
percentiles = [r[0] for r in results]
correlations = [r[1] for r in results]

# พล็อตกราฟ (กลับแกน X เพื่อให้เห็นความเข้มงวดจากน้อยไปมาก)
plt.plot(percentiles, correlations, marker='o', markersize=8, linestyle='-', color='crimson', linewidth=2.5)
plt.gca().invert_xaxis() 

plt.title("The Malinvestment Curve (Berkson's Paradox in HR Screening)", fontsize=14, fontweight='bold')
plt.xlabel("HR Strictness (Top % Selected)", fontsize=12)
plt.ylabel("Correlation (True Skill vs Resume Hacking)", fontsize=12)

# วาดเส้นฐาน (Correlation = 0) คือจุดที่ความสามารถ 2 ด้านเป็นอิสระต่อกัน
plt.axhline(0, color='black', linewidth=1.5, linestyle='--')
plt.annotate('Natural State (Independent)', xy=(95, 0.02), color='black')

plt.grid(True, linestyle=':', alpha=0.7)
plt.show()
```

```python
# 1. สร้างกลุ่มประชากรจำลอง (ผู้สมัคร 10,000 คน)
np.random.seed(42)
N = 10000

true_skill = np.random.normal(50, 15, N)
resume_skill = np.random.normal(50, 15, N)

# 2. สร้างเกณฑ์คัดกรอง (The Collider)
threshold = 150
total_score = true_skill + resume_skill
is_selected = total_score > threshold

skill_selected = true_skill[is_selected]
resume_selected = resume_skill[is_selected]

skill_rejected = true_skill[~is_selected]
resume_rejected = resume_skill[~is_selected]

slope, intercept, r_value, p_value, std_err = stats.linregress(skill_selected, resume_selected)
line_x = np.array([min(skill_selected), max(skill_selected)])
line_y = intercept + slope * line_x

# 3. สร้างกราฟด้วย Matplotlib (เวอร์ชันปรับปรุง)
plt.figure(figsize=(10, 10)) # ปรับขนาดให้เป็นสี่เหลี่ยมจัตุรัส

plt.scatter(skill_rejected, resume_rejected, color='lightgray', alpha=0.3, s=10, label='Rejected (General Pool)')
plt.scatter(skill_selected, resume_selected, color='crimson', alpha=0.6, s=15, label='Selected (Interviewees)')

# วาดเส้น Trendline
plt.plot(line_x, line_y, color='black', linewidth=3, linestyle='--', 
         label=f'Apparent Trend (Slope: {slope:.2f}, Corr: {r_value:.2f})')

# วาดเส้นกำแพง HR (Threshold Line) เพื่อให้กราฟดูสมเหตุสมผล
# สมการคือ X + Y = 120 ย้ายข้างจะได้ Y = 120 - X
x_thresh = np.array([10, 110])
y_thresh = threshold - x_thresh
plt.plot(x_thresh, y_thresh, color='blue', linewidth=2.5, linestyle='-', label='HR Threshold ($X + Y = 120$)')

plt.title("Berkson's Paradox: Negative Correlation in HR Screening", fontsize=14, fontweight='bold')
plt.xlabel("True Engineering Skill (V)", fontsize=12)
plt.ylabel("Resume Hacking Skill (\u03B5)", fontsize=12)

# บังคับสัดส่วนแกน X และ Y ให้เท่ากัน (1:1) เพื่อไม่ให้กราฟหลอกตา
plt.axis('equal') 
plt.xlim(0, 100)
plt.ylim(0, 100)

plt.axvline(50, color='gray', linestyle=':', alpha=0.5)
plt.axhline(50, color='gray', linestyle=':', alpha=0.5)

plt.legend(loc='upper right', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
# 1. สร้างประชากร (Uniform Distribution) กระจายตัวในกรอบ [0, 1]
np.random.seed(42)
N = 1000
handsome = np.random.uniform(0, 1, N)
nice = np.random.uniform(0, 1, N)
total_score = handsome + nice

# 2. กำหนดเงื่อนไขเส้นตัด (Thresholds) เหมือนในรูป
# เส้นสีเขียว: x + y = 1.4
# เส้นสีแดง: x + y = 1.8
mask_black = total_score <= 1.4
mask_green = (total_score > 1.4) & (total_score <= 1.8)
mask_red = total_score > 1.8

# 3. คำนวณความสัมพันธ์ (Correlation) ของแต่ละกลุ่ม
corr_all, _ = pearsonr(handsome, nice)
corr_black, _ = pearsonr(handsome[mask_black], nice[mask_black])
corr_green, _ = pearsonr(handsome[mask_green], nice[mask_green])
corr_red, _ = pearsonr(handsome[mask_red], nice[mask_red])

# ปรินต์ผลลัพธ์เพื่อพิสูจน์ Paradox
print(f"Correlation ประชากรทั้งหมด (ก่อนคัดกรอง): {corr_all:.4f}")
print(f"Correlation กลุ่มสีดำ (ต่ำกว่าเกณฑ์ 1.4): {corr_black:.4f}")
print(f"Correlation กลุ่มสีเขียว (ผ่านเกณฑ์ 1.4 แต่ไม่ถึง 1.8): {corr_green:.4f}")
print(f"Correlation กลุ่มสีแดง (Top Tier เกิน 1.8): {corr_red:.4f}")

# 4. พล็อตกราฟเลียนแบบต้นฉบับ
plt.figure(figsize=(7, 7))

# พล็อตจุดทั้ง 3 กลุ่ม
plt.scatter(handsome[mask_black], nice[mask_black], facecolors='none', edgecolors='k', s=25, label=f'Black (Corr: {corr_black:.2f})')
plt.scatter(handsome[mask_green], nice[mask_green], color='green', s=25, label=f'Green (Corr: {corr_green:.2f})')
plt.scatter(handsome[mask_red], nice[mask_red], color='red', s=25, label=f'Red (Corr: {corr_red:.2f})')

# พล็อตเส้นทแยงมุม
x_line = np.linspace(0, 1, 100)
plt.plot(x_line, 1.4 - x_line, color='green', linewidth=3)
plt.plot(x_line, 1.8 - x_line, color='red', linewidth=3)

# ตกแต่งกราฟให้เหมือนต้นฉบับ
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.xlabel("Handsome -->", fontsize=16)
plt.ylabel("Nice -->", fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=14)

# เพิ่ม Legend เพื่อโชว์ค่า Corr ในกราฟด้วย
plt.legend(loc='lower left')
plt.show()
```

```python
# 1. สร้างประชากร (Uniform Distribution) กระจายตัวในกรอบ [0, 1]
np.random.seed(42)
N = 1000
handsome = np.random.uniform(0, 1, N)
charm = np.random.uniform(0, 1, N) # ปรับแกน Y เป็นทักษะคารม/ความเจ้าชู้
total_score = handsome + charm

# 2. กำหนดเงื่อนไขเส้นตัด (Thresholds)
mask_black = total_score <= 1.4
mask_green = (total_score > 1.4) & (total_score <= 1.8)
mask_red = total_score > 1.8

# 3. คำนวณความสัมพันธ์ (Correlation) ของแต่ละกลุ่ม
corr_all, _ = pearsonr(handsome, charm)
corr_black, _ = pearsonr(handsome[mask_black], charm[mask_black])
corr_green, _ = pearsonr(handsome[mask_green], charm[mask_green])
corr_red, _ = pearsonr(handsome[mask_red], charm[mask_red])

# ปรินต์ผลลัพธ์
print(f"Correlation ประชากรทั้งหมด (ก่อนคัดกรอง): {corr_all:.4f}")
print(f"Correlation กลุ่มสีดำ (ต่ำกว่าเกณฑ์ 1.4): {corr_black:.4f}")
print(f"Correlation กลุ่มสีเขียว (ผ่านเกณฑ์ 1.4 แต่ไม่ถึง 1.8): {corr_green:.4f}")
print(f"Correlation กลุ่มสีแดง (Top Tier เกิน 1.8): {corr_red:.4f}")

# 4. พล็อตกราฟ
plt.figure(figsize=(8, 8))

# พล็อตจุดทั้ง 3 กลุ่ม
plt.scatter(handsome[mask_black], charm[mask_black], facecolors='none', edgecolors='k', s=25, label=f'Rejected (Corr: {corr_black:.2f})')
plt.scatter(handsome[mask_green], charm[mask_green], color='green', s=25, label=f'Accepted Pool (Corr: {corr_green:.2f})')
plt.scatter(handsome[mask_red], charm[mask_red], color='red', s=25, label=f'Top Tier (Corr: {corr_red:.2f})')

# พล็อตเส้นทแยงมุม
x_line = np.linspace(0, 1, 100)
plt.plot(x_line, 1.4 - x_line, color='green', linewidth=3)
plt.plot(x_line, 1.8 - x_line, color='red', linewidth=3)

# # เพิ่มไฮไลต์โซน "Ugly Playboy" (หน้าตาไม่ดีแต่รอดมาได้เพราะเจ้าชู้จัด)
# # ในกราฟ Uniform [0,1] คนที่หล่อน้อยที่สุดที่จะผ่านเส้น 1.4 ได้คือหล่อ=0.4 (เพราะคารมตันที่ 1.0)
# rect = patches.Rectangle((0.4, 0.8), 0.25, 0.2, linewidth=2, edgecolor='darkorange', facecolor='none', linestyle='--', label='Ugly Playboy Zone')
# plt.gca().add_patch(rect)

# ตกแต่งกราฟ
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.xlabel("Handsome -->", fontsize=16)
plt.ylabel("Charm (Playboy Skill) -->", fontsize=16)
plt.title("Berkson's Paradox: Dating Market (Uniform Distribution)", fontsize=16, fontweight='bold')
plt.tick_params(axis='both', which='major', labelsize=14)

# เพิ่ม Legend
plt.legend(loc='lower left', fontsize=12)
plt.show()
```

## การทำลาย Berkson's Paradox ด้วยฮิวริสติก Take-The-Best (TTB) ของ Gigerenzer

อัลกอริทึมคัดกรองเรซูเม่ (ATS) ของ HR หรือแม้แต่โมเดล Machine Learning ส่วนใหญ่บนโลก ล้วนตกหลุมพรางของ Berkson's Paradox เพราะพวกมันถูกสร้างมาบนพื้นฐานของ **โมเดลแบบชดเชยกันได้ (Compensatory Models)** ในขณะที่ฮิวริสติก **Take-The-Best (TTB)** ของ Gerd Gigerenzer ใช้แนวทางที่สวนทางอย่างสิ้นเชิง คือ **โมเดลแบบไม่ชดเชย (Non-compensatory Models)** 

### 1. ทำไมระบบเดิมถึงพัง? (The Compensatory Trap)
ในระบบเดิม (เช่น Multiple Regression หรือ Weighted Scoring) คะแนนรวม $S$ เกิดจากการนำทักษะจริง ($X$) และทักษะการทำเรซูเม่ ($Y$) มาคูณน้ำหนักแล้วบวกกัน:
$$S = w_1X + w_2Y$$

ความพังพินาศเกิดจากการที่สมการนี้อนุญาตให้ผู้สมัคร **"ใช้ $Y$ มาชดเชย $X$ ที่ขาดหายไปได้"** เมื่อ HR ตั้งเกณฑ์รับคนเกิดเป็นกำแพง (Collider) มันจึงบีบให้ตัวแปรอิสระสองตัวเกิดความสัมพันธ์เชิงลบ (Negative Correlation) ทันที

### 2. กลไกของ Take-The-Best (The Antidote)
TTB ปฏิเสธการนำตัวแปรมาบวกกันโดยสิ้นเชิง แต่จะประเมินข้อมูลเป็นลำดับชั้น (Lexicographic) โดยมีกฎเหล็ก 3 ข้อ:

1. **Search Rule (กฎการค้นหา):** เรียงลำดับตัวชี้วัด (Cues) ตามความแม่นยำ (Validity) จากมากไปน้อย เช่น ให้ $X$ (ทดสอบเขียนสถาปัตยกรรมระบบจริง) เป็น Cue ที่ 1 และ $Y$ (คีย์เวิร์ดในเรซูเม่) เป็น Cue ที่ 2
2. **Stopping Rule (กฎการหยุด):** เปรียบเทียบผู้สมัครทีละ Cue **หากเจอจุดที่แตกต่างกัน ให้ "หยุดค้นหา" ทันที** ไม่ต้องดู Cue ถัดไปอีก
3. **Decision Rule (กฎการตัดสินใจ):** ตัดสินใจเลือกคนที่มีค่าใน Cue นั้นดีกว่า ทันที!

### 3. บทพิสูจน์: TTB ทำลาย Collider Bias ได้อย่างไร?
สมมติว่ามีผู้สมัคร 2 คน คือ A และ B
* **A:** ทักษะจริงสูงมาก ($X=90$) แต่ทำเรซูเม่ห่วยแตก ($Y=10$)
* **B:** ทักษะจริงต่ำ ($X=30$) แต่ใช้ AI เขียนเรซูเม่จนสมบูรณ์แบบ ($Y=95$)

**หากใช้ระบบ HR แบบดั้งเดิม (คะแนนรวม):**
หากน้ำหนักเท่ากัน A ได้ 100 คะแนน ส่วน B ได้ 125 คะแนน ระบบจะเลือก B ทันที และเมื่อทำแบบนี้ซ้ำๆ องค์กรจะเต็มไปด้วยคนแบบ B (นี่คือจุดกำเนิดของ Negative Correlation และ Malinvestment)

**หากใช้ Take-The-Best:**
* **ขั้นตอนที่ 1:** เปรียบเทียบ Cue ที่ 1 ($X$ - ทักษะจริง)
* **ขั้นตอนที่ 2:** พบว่า $X_A (90) > X_B (30)$ ผู้สมัคร A ชนะใน Cue นี้
* **ขั้นตอนที่ 3:** **ระบบจะหยุดทำงานและรับ A เข้าทำงานทันที โดยไม่แม้แต่จะชายตามองตัวแปร $Y$**

**บทสรุปทางสถิติ:**
ด้วยกลไก Stopping Rule ของ TTB ทักษะการทำเรซูเม่ ($Y$) จะถูกตัดขาดออกจากสมการการตัดสินใจโดยสิ้นเชิงเมื่อทักษะจริงชี้ขาดไปแล้ว มันจึงไม่มีโอกาสได้เข้ามา "ชดเชย" หรือรวมตัวกับ $X$ เพื่อสร้างจุดตัด (Collider) อีกต่อไป เมื่อไม่มี Collider ความสัมพันธ์เชิงลบระหว่าง $X$ กับ $Y$ ก็ไม่เกิด Berkson's Paradox จึงถูกทำลายลงอย่างสมบูรณ์แบบ

การประยุกต์ใช้ TTB หรือ Fast-and-Frugal Trees ในองค์กร คือการยอมรับว่าโลกนี้มีความไม่แน่นอนสูง การพยายามเอาทุกตัวแปรมาคำนวณรวมกันคือความโง่เขลา (Overfitting) การใช้ตัวชี้วัดที่ทรงพลังเพียงตัวเดียวแล้วตัดจบต่างหาก คือความแข็งแกร่ง (Robustness) ที่แท้จริง
