```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm
from pandas_datareader import data as dr


from scipy.stats import chi2
from matplotlib import patches
```

```python
ticker = 'GOOGL'  #GOOGLE stock ticker
data = pd.DataFrame(columns = [ticker])
data[ticker] = dr.DataReader(ticker, data_source = 'yahoo', start = '2016-1-1', end = '2020-12-31')['Adj Close']
data.head()
```

```python
returns = data.pct_change()
returns.dropna(inplace = True)
```

```python
sns.distplot(returns, hist = True)
plt.show()
```

```python
l = norm.ppf(0.10)
u = norm.ppf(0.85)
```

```python
l
```

```python
u
```

```python
mean = returns.mean()
stdev = returns.std()
#np.random.seed(0)
n = np.random.normal(size = (30,10))
rows = n.shape[0]
cols = n.shape[1]
```

```python
for i in range(0,rows) :
    for j in range(0,cols) :
        if n[i][j] > u :
            n[i][j] = u       #sets upper limit
        elif n[i][j] < l :
            n[i][j] = l     #sets lower limit
        else :
            n[i][j] = n[i][j]
        n[i][j] = (stdev * n[i][j]) + mean
```

```python
s = data.iloc[-1]
pred = np.ones_like(n)
pred[0] = s         #sets beginning point of simulations
```

```python
for i in range(1,30) :
    pred[i] = pred[(i-1)] * (1 + n[(i-1)])
```

```python
plt.figure(figsize = (7,4))
plt.title('{} SIMULATIONS'.format(ticker))    
for i in range(0,10) :
    plt.plot(pred[:, i])
plt.show()
```

# Outlier Detect

```python
df = pred[:, :2]
```

```python
covariance  = np.cov(df , rowvar=False)
```

```python
covariance
```

```python
covariance_pm1 = np.linalg.inv(covariance)
```

```python
centerpoint = np.mean(df , axis=0)
```

```python
centerpoint
```

```python
distances = []
for i, val in enumerate(df):
    distance = (val-centerpoint).T @  covariance_pm1 @ (val-centerpoint)
    distances.append(distance)
distances = np.array(distances)
```

```python
distances
```

```python
cutoff = chi2.ppf(0.95, df.shape[1])
```

```python
cutoff
```

```python
df.shape[1]
```

```python
outlierIndexes = np.where(distances > cutoff )
```

```python
outlierIndexes
```

```python
chi2.cdf(distances, 2)
```

```python
returns = data.pct_change().dropna().values
```

```python
S_init = returns[0]
dt = 1
T = data.shape[0]
N = int(T / dt)
t = np.arange(1, int(N) + 1)
returns = returns.flatten()
mu = returns.mean()
sigma = returns.std()
```

```python
# Scen size, time period, num of assets
```

```python
scen_size = 100
b = np.random.normal(0, 1, size=(scen_size, N))
w = b.cumsum(axis=1)
drift = (mu - 0.5 * sigma**2) * t
diffusion = w * sigma
```

```python
S = S_init * np.exp(drift + diffusion)
S_history = np.hstack([S_init * np.ones((scen_size, 1)), S])
```

# Monte carlo Portfolio 

```python
coins = pd.read_csv('stable_coin_27days.csv', index_col=0)
```

```python
coins.head()
```

```python
returns_ret = coins.pct_change().dropna().values
```

```python
returns_ret.shape
```

```python
mu = returns_ret.mean(axis=0)
sigma = returns_ret.std(axis=0)
cov = np.cov(returns_ret, rowvar=False)
S_init = returns_ret[0]
```

```python
T = coins.shape[0]
num_coins = coins.shape[1]
scen_size = 1000
dt = 1
N = int(T / dt)
t = np.array([np.arange(1, int(N) + 1)] * num_coins).T
A = np.linalg.cholesky(cov)
b = np.random.normal(0, 1, size=(scen_size, N, num_coins))
w = b.cumsum(axis=1)

Z = np.random.normal(0, 1, size=(scen_size, N, num_coins))
drift = (mu - 0.5 * sigma**2) * t
diffusion = w @ A

S = S_init * np.exp(drift + diffusion)
S_history = np.array([np.vstack([S_init, s]) for s in S])
```

```python
Weight = np.ones(num_coins) / num_coins
```

```python
Weight
```

```python
Earn = S_history @ Weight
```

```python
for e in Earn:       
    plt.plot(e)
plt.show()
```

```python
Earn[:, -1].mean()
```
