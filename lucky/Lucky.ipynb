```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
x = np.linspace(-5, 5, 100)
lucky_function = lambda x: (np.exp(x) - 1) / (np.e) - 1
plt.plot(x, lucky_function(x))
plt.show()
```

```python
np.random.seed(42)
random_x = np.random.randn(365, 100) * 3
```

```python
walk_without_lucky = np.cumsum(random_x, axis=0)

plt.plot(walk_without_lucky);
```

```python
walk_with_lucky = np.cumsum(lucky_function(random_x), axis=0)

plt.plot(walk_with_lucky);
```

```python
np.mean(walk_with_lucky[-1, :])
```

```python
np.mean(walk_without_lucky[-1, :])
```
